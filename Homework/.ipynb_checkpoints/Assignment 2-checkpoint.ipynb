```python
def leastsq_function(x,y):
    # 单变量时的最小二乘法
    n = len(x)
    sumX = x.sum(axis = 0)
    sumY = y.sum(axis = 0)
    sumXY = x*y.sum(axis = 0)
    sumXX = x*x.sum(axis = 0)
    a = (sumXY - (1/n) * (sumX * sumY)) / (sumXX - (1/n) * sumX * sumX)
    b = sumY/n - a * sumX/n
    return a, b
```

```python
def multi_leastsq_function(x,y):
    A = np.hstack([x, np.ones((len(x),1))])
    W = np.linalg.lstsq(A, y, rcond=None)[0]
    a = W[:-1]
    b = W[-1]
    return a,b
```
